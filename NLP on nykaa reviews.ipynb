```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
import re
import emoji
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,LancasterStemmer,SnowballStemmer,WordNetLemmatizer
import sklearn
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
df=df.dropna(subset=["review_text"])
```

```python
df.isnull().sum()
```

```python
stp=stopwords.words("english")
stp.remove("not")
```

```python
new_r=[]
for sent in df["review_text"]:
    l=[]
    ns=str(sent).lower()
    ns=emoji.demojize(ns)
    ns=re.sub("[<.*?>]","",ns)
    ns=re.sub("https?//\S*|www\S*","",ns)
    ns=re.sub("[^a-z ]","",ns)
    for word in word_tokenize(ns):
        if word in stp:
            pass
        else:
            stems=WordNetLemmatizer().lemmatize(word)
            l.append(stems)
    new_r.append(" ".join(l))
            
```

```python
df1=pd.DataFrame({"cleaned_review":new_r})
```

```python
df1
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
new_data=cv.fit_transform(df1["cleaned_review"])
new_data.toarray()
```

```python
pd.DataFrame(new_data.toarray(),columns=sorted(cv.vocabulary_.keys()))
```
